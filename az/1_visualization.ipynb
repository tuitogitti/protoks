```python
import numpy as np # matematiikkakirjasto
import pandas as pd # datan käsittelykirjasto
import matplotlib.pyplot as plt # graafien piirtokirjasto
```

```python
df1 = pd.read_csv('data/imu_skdata1_031125.csv') # tehoton sauvakävely datanäyte
```

```python
df2 = pd.read_csv('data/imu_skdata2_031125.csv') # tehokas sauvakävely datanäyte
```

```python
df3 = pd.read_csv('data/imu_skdata3_031125.csv') # tehoton ja tehokas sauvakävely samassa datanäytteessä
```

```python
df1.head()
```

```python
acc_df1 = df1.iloc[:,[0,1,2,3]] # otetaan accelerometridata erilleen
```

```python
acc_df1.head()
```

```python
acc_df2 = df2.iloc[:,[0,1,2,3]] # otetaan accelerometridata erilleen
```

```python
acc_df3 = df3.iloc[:,[0,1,2,3]] # otetaan accelerometridata erilleen
```

```python
gyr_df1 = df1.iloc[:,[0,4,5,6]] # otetaan gyroskooppidata erilleen
```

```python
gyr_df1.head()
```

```python
gyr_df2 = df2.iloc[:,[0,4,5,6]] # otetaan gyroskooppidata erilleen
```

```python
gyr_df3 = df3.iloc[:,[0,4,5,6]] # otetaan gyroskooppidata erilleen
```

```python
# timestamp indeksiksi
acc_df1.set_index('Timestamp', inplace=True)
```

```python
# timestamp indeksiksi
acc_df2.set_index('Timestamp', inplace=True)
```

```python
# timestamp indeksiksi
acc_df3.set_index('Timestamp', inplace=True)
```

```python
# timestamp indeksiksi
gyr_df1.set_index('Timestamp', inplace=True)
```

```python
# timestamp indeksiksi
gyr_df2.set_index('Timestamp', inplace=True)
```

```python
# timestamp indeksiksi
gyr_df3.set_index('Timestamp', inplace=True)
```

```python
# Kiihtyvyys ajan funktiona
plt.figure(figsize=(12, 6))
plt.plot(acc_df1.index, acc_df1['AccX'], label='x')
plt.plot(acc_df1.index, acc_df1['AccY'], label='y')
plt.plot(acc_df1.index, acc_df1['AccZ'], label='z')
plt.title('Tehoton sauvakävelytekniikka: kiihtyvyys')
plt.xlabel('Time')
plt.ylabel('Acceleration (m/s^2)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Kiihtyvyys ajan funktiona
plt.figure(figsize=(12, 6))
plt.plot(acc_df2.index, acc_df2['AccX'], label='x')
plt.plot(acc_df2.index, acc_df2['AccY'], label='y')
plt.plot(acc_df2.index, acc_df2['AccZ'], label='z')
plt.title('Tehokas sauvakävelytekniikka: kiihtyvyys')
plt.xlabel('Time')
plt.ylabel('Acceleration (m/s^2)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Kiihtyvyys ajan funktiona
plt.figure(figsize=(12, 6))
plt.plot(acc_df3.index, acc_df3['AccX'], label='x')
plt.plot(acc_df3.index, acc_df3['AccY'], label='y')
plt.plot(acc_df3.index, acc_df3['AccZ'], label='z')
plt.title('Tehoton ja tehokas sauvakävelytekniikka rinnakkain: kiihtyvyys')
plt.xlabel('Time')
plt.ylabel('Acceleration (m/s^2)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Kulmarotaatio ajan funktiona
plt.figure(figsize=(12, 6))
plt.plot(gyr_df1.index, gyr_df1['GyroX'], label='x')
plt.plot(gyr_df1.index, gyr_df1['GyroY'], label='y')
plt.plot(gyr_df1.index, gyr_df1['GyroZ'], label='z')
plt.title("Tehoton sauvakävelytekniikka: rotaatio")
plt.xlabel('Time')
plt.ylabel('Rotation (rad/s)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Kulmarotaatio ajan funktiona
plt.figure(figsize=(12, 6))
plt.plot(gyr_df2.index, gyr_df2['GyroX'], label='x')
plt.plot(gyr_df2.index, gyr_df2['GyroY'], label='y')
plt.plot(gyr_df2.index, gyr_df2['GyroZ'], label='z')
plt.title("Tehokas sauvakävelytekniikka: rotaatio")
plt.xlabel('Time')
plt.ylabel('Rotation (rad/s)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Kulmarotaatio ajan funktiona
plt.figure(figsize=(12, 6))
plt.plot(gyr_df3.index, gyr_df3['GyroX'], label='x')
plt.plot(gyr_df3.index, gyr_df3['GyroY'], label='y')
plt.plot(gyr_df3.index, gyr_df3['GyroZ'], label='z')
plt.title("Tehoton ja tehokas sauvakävelytekniikka rinnakkain: rotaatio")
plt.xlabel('Time')
plt.ylabel('Rotation (rad/s)')
plt.legend()
plt.grid(True)
plt.show()
```
